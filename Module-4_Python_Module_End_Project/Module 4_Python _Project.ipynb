```python
import pandas as pd

# Assuming your imported data is in a DataFrame called 'data'
# Example: data = pd.read_excel('python project work.xlsx')

# Print out the columns to verify their names
print("Columns in the DataFrame:", data.columns)

# Function to calculate IQR
def calculate_iqr(column):
    Q1 = column.quantile(0.25)
    Q3 = column.quantile(0.75)
    IQR = Q3 - Q1
    return IQR

# Selecting only numeric columns
numeric_columns = data.select_dtypes(include='number')

# Calculating IQR for each numeric column
iqr_values = numeric_columns.apply(calculate_iqr)

# Displaying the IQR values
print(iqr_values)
```

```python
##Explanation of IQR for Each Column:
#Number: The IQR of 20.00 means the middle 50% of the data for the 'Number' column is spread out over 20 units. This indicates moderate variability within the 'Number' column.

#Age: The IQR of 6.00 means the middle 50% of the data for the 'Age' column is spread over 6 years. This shows a fairly tight range of ages within the dataset.

#Height: The IQR of 16.00 indicates that the middle 50% of the 'Height' values are spread over 16 units. This suggests moderate variability in heights.

Weight: The IQR of 40.00 implies that the middle 50% of the 'Weight' values are spread over 40 units. This shows greater variability in weights compared to age and height.

Salary: The IQR of 5,223,402.50 indicates a very high variability in salaries within the middle 50% of the dataset. This suggests that salaries are widely spread out and could include very different amounts.

Significance of IQR:
Moderate IQR values (like those for 'Number', 'Age', and 'Height') indicate that data is moderately spread out.

High IQR values (like those for 'Weight' and especially 'Salary') indicate significant variability in the data. This is particularly important for salary, as it shows a wide range in the middle 50%, highlighting potential disparities or diverse salary structures.
```

## 1. Determine the distribution of employees across each team and calculate the percentage split relative to the total number of employees. (2 marks)


```python
# Count the number of employees in each team
team_distribution = data['Team'].value_counts().reset_index()

# Rename the columns for clarity
team_distribution.columns = ['Team', 'count']

# Calculate the total number of employees
total_employees = data.shape[0]

# Calculate the percentage split
team_distribution['percentage'] = (team_distribution['count'] / total_employees) * 100

# Reset index to start from 1
team_distribution.index = team_distribution.index + 1

print("Team distribution with index number starting from 1:\n", team_distribution)
```

## 2. Segregate employees based on their positions within the company. (2 marks)


```python
# Group employees by their position
position_groups = data.groupby('Position').size().reset_index(name='count')
position_groups.index = position_groups.index + 1

print("Employees segregated based on their positions:\n", position_groups)
```

```python
# Find the maximum age in the 'age' column
max_age = data['Age'].max()

print("The maximum age is:", max_age)
```

## 3. Identify the predominant age group among employees. (2 marks)


```python
# Define age groups
bins = [0, 10, 20, 30, 40]
labels = ['0-10', '11-20', '21-30', '31-40']

# Create a new column for age groups
data['Age_group'] = pd.cut(data['Age'], bins=bins, labels=labels, right=True)

# Count the number of employees in each age group
age_group_distribution = data['Age_group'].value_counts().reset_index()
age_group_distribution.columns = ['Age_group', 'count']

# Identify the predominant age group
predominant_age_group = age_group_distribution.loc[age_group_distribution['count'].idxmax()]

print("Age group distribution:\n", age_group_distribution)
print("\nPredominant age group:\n", predominant_age_group)
```

## 4. Discover which team and position have the highest salary expenditure. (2 marks)


```python
# Group by team and position and calculate the total salary expenditure
salary_expenditure = data.groupby(['Team', 'Position'])['Salary'].sum().reset_index()

# Set display option to show floating-point numbers with two decimal places
pd.options.display.float_format = '{:.2f}'.format

# Identify the team and position with the highest salary expenditure
max_expenditure = salary_expenditure.loc[salary_expenditure['Salary'].idxmax()]

print("Salary expenditure by Team and Position:\n", salary_expenditure)
print("\nTeam and position with the highest salary expenditure:\n", max_expenditure)
```

## 5. Investigate if there's any correlation between age and salary, and represent it visually. (2 marks)



```python
# Calculate the correlation between age and salary
correlation = data['Age'].corr(data['Salary'])
print(f"Correlation between age and salary: {correlation:.2f}")
```
