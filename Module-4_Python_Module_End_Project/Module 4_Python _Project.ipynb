```python
import pandas as pd

# Assuming your imported data is in a DataFrame called 'data'
# Example: data = pd.read_excel('python project work.xlsx')

# Print out the columns to verify their names
print("Columns in the DataFrame:", data.columns)

# Function to calculate IQR
def calculate_iqr(column):
    Q1 = column.quantile(0.25)
    Q3 = column.quantile(0.75)
    IQR = Q3 - Q1
    return IQR

# Selecting only numeric columns
numeric_columns = data.select_dtypes(include='number')

# Calculating IQR for each numeric column
iqr_values = numeric_columns.apply(calculate_iqr)

# Displaying the IQR values
print(iqr_values)
```

```python
##Explanation of IQR for Each Column:
#Number: The IQR of 20.00 means the middle 50% of the data for the 'Number' column is spread out over 20 units.
#This indicates moderate variability within the 'Number' column.

#Age: The IQR of 6.00 means the middle 50% of the data for the 'Age' column is spread over 6 years.
#This shows a fairly tight range of ages within the dataset.

#Height: The IQR of 16.00 indicates that the middle 50% of the 'Height' values are spread over 16 units. 
#This suggests moderate variability in heights.

#Weight: The IQR of 40.00 implies that the middle 50% of the 'Weight' values are spread over 40 units. 
#This shows greater variability in weights compared to age and height.

#Salary: The IQR of 5,223,402.50 indicates a very high variability in salaries within the middle 50% of the dataset. 
#This suggests that salaries are widely spread out and could include very different amounts.

#Significance of IQR:
#Moderate IQR values (like those for 'Number', 'Age', and 'Height') indicate that data is moderately spread out.

#High IQR values (like those for 'Weight' and especially 'Salary') indicate significant variability in the data. 
#This is particularly important for salary, as it shows a wide range in the middle 50%, 
#highlighting potential disparities or diverse salary structures.
```

## 1. Determine the distribution of employees across each team and calculate the percentage split relative to the total number of employees. (2 marks)


```python
# Count the number of employees in each team
team_distribution = data['Team'].value_counts().reset_index()

# Rename the columns for clarity
team_distribution.columns = ['Team', 'count']

# Calculate the total number of employees
total_employees = data.shape[0]

# Calculate the percentage split
team_distribution['percentage'] = (team_distribution['count'] / total_employees) * 100

# Reset index to start from 1
team_distribution.index = team_distribution.index + 1

print("Team distribution with index number starting from 1:\n", team_distribution)
```

```python
import matplotlib.pyplot as plt

# Assuming 'data' is your DataFrame containing the employee data

# Calculate the distribution of employees across teams
team_distribution = data['Team'].value_counts().reset_index()

# Rename the columns for clarity
team_distribution.columns = ['Team', 'count']

# Calculate the total number of employees
total_employees = data.shape[0]

# Calculate the percentage split
team_distribution['percentage'] = (team_distribution['count'] / total_employees) * 100

# Find indices of max and min values
max_idx = team_distribution['percentage'].idxmax()
min_idx = team_distribution['percentage'].idxmin()

# Create explode array
explode = [0.1 if i in [max_idx, min_idx] else 0 for i in range(len(team_distribution))]

# Plot pie chart with explode
plt.figure(figsize=(12, 12))
plt.pie(team_distribution['percentage'], labels=team_distribution['Team'], autopct='%1.1f%%', startangle=140, explode=explode)
plt.title('Distribution of Employees Across Teams')
plt.show()
```

## 3. Identify the predominant age group among employees. (2 marks)


```python
# Define age groups
bins = [0, 10, 20, 30, 40]
labels = ['0-10', '11-20', '21-30', '31-40']

# Create a new column for age groups
data['Age_group'] = pd.cut(data['Age'], bins=bins, labels=labels, right=True)

# Count the number of employees in each age group
age_group_distribution = data['Age_group'].value_counts().reset_index()
age_group_distribution.columns = ['Age_group', 'count']

# Identify the predominant age group
predominant_age_group = age_group_distribution.loc[age_group_distribution['count'].idxmax()]

print("Age group distribution:\n", age_group_distribution)
print("\nPredominant age group:\n", predominant_age_group)
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Assuming your DataFrame is called 'age_group_distribution'

# Updated bar plot code
plt.figure(figsize=(10, 6))
sns.set(style="whitegrid")
bar_plot = sns.barplot(y='Age_group', x='count', data=age_group_distribution, palette='viridis', hue='Age_group', dodge=False)
plt.legend([], [], frameon=False)

# Add titles and labels
plt.title('Age Group Distribution')
plt.xlabel('Count')
plt.ylabel('Age Group')

# Get the bars
bars = bar_plot.patches

# Add text on top of each bar
for bar in bars:
    width = bar.get_width()
    plt.text(bar.get_x() + width + 6, bar.get_y() + bar.get_height() / 2, f'{int(width)}', ha='center', va='center', color='black')

# Show the plot
plt.show()
```

### The horizontal bar chart effectively displays the distribution of employees across different age groups,
### with the 21-30 age group having the highest count, followed by 31-40 and 11-20 age groups,
### indicating a younger demographic within the organization. The chart features a light gray background and teal bars,
### with employee counts displayed next to each bar for clarity. This visual representation helps in understanding age diversity,
### recruitment strategies, and workforce planning.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Provided data
data = pd.read_excel('python project work.xlsx')

# Group by team and position and calculate the total salary expenditure
salary_expenditure = data.groupby(['Team', 'Position'])['Salary'].sum().reset_index()

# Identify the team and position with the highest salary expenditure
max_expenditure = salary_expenditure.loc[salary_expenditure['Salary'].idxmax()]

# Create a bar plot to visualize salary expenditure by team and position
plt.figure(figsize=(12, 8))
bar_plot = sns.barplot(x='Salary', y='Team', hue='Position', data=salary_expenditure, palette='viridis')
plt.legend(title='Position')

# Add titles and labels
plt.title('Salary Expenditure by Team and Position')
plt.xlabel('Salary')
plt.ylabel('Team')

# Highlight the team and position with the highest salary expenditure
plt.text(max_expenditure['Salary'], max_expenditure.name, f'{max_expenditure["Team"]} - {max_expenditure["Position"]}: ${max_expenditure["Salary"]:.2f}', 
         ha='center', va='bottom', color='red', weight='bold')

# Show the plot
plt.show()
```

## 5. Investigate if there's any correlation between age and salary, and represent it visually. (2 marks)



```python
# Calculate the correlation between age and salary
correlation = data['Age'].corr(data['Salary'])
print(f"Correlation between age and salary: {correlation:.2f}")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Create a scatter plot with a regression line
plt.figure(figsize=(10, 6))
sns.regplot(x='Age', y='Salary', data=data, scatter_kws={'color': 'blue'}, line_kws={'color': 'red'})

# Add titles and labels
plt.title('Scatter Plot with Regression Line: Age vs. Salary')
plt.xlabel('Age')
plt.ylabel('Salary')

# Show the plot
plt.show()
```

### Weak Correlation: The near-horizontal red line and the spread-out data points indicate that 'Age' has little to no effect on 'Salary'.
### The correlation coefficient of 0.21 aligns with this observation, highlighting a weak positive correlation.
### Implications: Other factors, such as experience, education, job role, and industry, likely have a more significant impact on 'Salary'
### than 'Age'. This finding emphasizes the importance of considering multiple variables when analyzing salary determinants.
